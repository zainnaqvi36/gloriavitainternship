```python
import random

# List of words for the game
word_list = ['python', 'programming', 'developer', 'algorithm', 'hiking', 'climbing', 'photography', 'digital', 'nomad']

# Function to select a random word
def choose_word(word_list):
    return random.choice(word_list)

# Function to display the masked word
def display_masked_word(word, correct_guesses):
    masked_word = ''.join([letter if letter in correct_guesses else '_' for letter in word])
    return masked_word

# Function to play the game
def play_game():
    print("Welcome to the Word Guessing Game!")
    
    # Difficulty level selection
    difficulty = input("Select difficulty level (easy/medium/hard): ").lower()
    
    # Set attempts based on difficulty
    if difficulty == 'easy':
        attempts = 10
    elif difficulty == 'medium':
        attempts = 7
    else:
        attempts = 5

    # Initialize game variables
    word = choose_word(word_list)
    correct_guesses = set()
    incorrect_guesses = set()
    
    print(f"\nYou have chosen {difficulty} difficulty. You have {attempts} attempts.")
    
    # Game loop
    while attempts > 0:
        # Display masked word
        masked_word = display_masked_word(word, correct_guesses)
        print("\nWord to guess: " + masked_word)
        
        # Player guesses a letter
        guess = input("Guess a letter: ").lower()
        
        if guess in correct_guesses or guess in incorrect_guesses:
            print(f"You've already guessed '{guess}'. Try a different letter.")
            continue
        
        # Check if the guess is in the word
        if guess in word:
            correct_guesses.add(guess)
            print(f"Good guess! '{guess}' is in the word.")
        else:
            incorrect_guesses.add(guess)
            attempts -= 1
            print(f"Wrong guess! '{guess}' is not in the word. Attempts remaining: {attempts}")
        
        # Check if the player has won
        if set(word) == correct_guesses:
            print("\nCongratulations! You've guessed the word correctly: " + word)
            return
        
    # If the player runs out of attempts
    print("\nSorry, you've run out of attempts. The word was: " + word)

# Function for main menu and game flow
def main():
    while True:
        play_game()
        replay = input("\nDo you want to play again? (yes/no): ").lower()
        if replay != 'yes':
            print("Thanks for playing! Goodbye!")
            break

if __name__ == '__main__':
    main()
```
